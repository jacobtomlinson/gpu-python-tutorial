Your challenge is to implement Fizz Buzz in CUDA Python with numba and calculate all instances of `Fizz`, `Buzz` and `Fizz Buzz` for the numbers between `1` and `500_000_000`.

```python
import numpy as np
from numba import cuda
```

**1. Create the input data array.** Try using `np.arange`.

```python
data = np.arange(1, 500_000_000)
data
```

**2. Create the output data array.**

```python
output = np.zeros_like(data)
output
```

**3. Calculate the number of threads and blocks.**

```python
threads = 128
blocks = len(data) + threads - 1 // 128
```

**4. Create the kernel.** _Tip: Our output array must be numeric so try using the values `1`, `2` and `3` to represent `Fizz`, `Buzz` and `Fizz Buzz`._

```python
@cuda.jit
def fizz_buzz(input_array, output_array):
    i = cuda.grid(1)
    n = input_array[i]
    if n % 3 == 0:
        output_array[i] = 1 # Fizz
    if n % 4 == 0:
        output_array[i] = 2 # Buzz
    if n % 12 == 0:
        output_array[i] = 3 # Fizz Buzz
```

**5. Run our kernel.**

```python
%%time
fizz_buzz[blocks, threads](data, output)
output[:12]
```
